```python
import os 
import glob
FOLDER = os.path.join(os.getcwd(), "preprocessing_parquet")

def join_path(filename: str):
    return os.path.join(FOLDER, filename)

def get_latest_file():
    list_of_files = glob.glob(f"{FOLDER}/*") # * means all if need specific format then *.csv
    return max(list_of_files, key=os.path.getctime)
```

```python
import pandas as pd
pd.read_parquet(get_latest_file())
```
